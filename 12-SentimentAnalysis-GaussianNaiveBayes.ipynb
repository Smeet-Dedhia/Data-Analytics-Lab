```python
def cont_to_exp(x):
    if type(x) is str:
        x = x.replace('\\','')
        for key in contractions:
            value = contractions[key]
            x = x.replace(key, value)
        return x
    else:
        return x
```

```python
punctuation = {".":"", ",":"", "'":"", "?":"", "!":"", ":":"", ";":"", "-":"", "_":"", "(":"", ")":"", 
       }
```

```python
def remove_punctuation(x):
    if type(x) is str:
        for key in punctuation:
            value = punctuation[key]
            x = x.replace(key, value)
        return x
    else:
        return x
```

```python
def remove_special_chars(text):
    special_chars = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    if type(text) is str:
        for p in special_chars:
            text = text.replace(p, '')
        return text
    else:
        return text
```

```python
df1["Review Text"] = df1["Review Text"].apply(lambda x:cont_to_exp(x))
df1["Review Text"]= df1["Review Text"].str.lower()
df1["Review Text"] = df1["Review Text"].apply(lambda x:remove_punctuation(x))
df1['Review Text'] = df1['Review Text'].apply(lambda x:remove_special_chars(x))
df1
```

### Q6] Independent & Dependent variables, 80-20 split

```python
from textblob import TextBlob
df1['polarity_title'] = df1['Title'].apply(lambda x: TextBlob(x).sentiment.polarity)
df1['polarity_review'] = df1['Review Text'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df1 = df1.drop(columns=['Clothing ID', 'Review Text', 'Title', 'Department Name', 'Division Name', 'Category'])
```

```python
x = df1.drop(columns=['Recommended IND'])

y = df1['Recommended IND']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
numerical_cols = [cname for cname in x.columns if x[cname].dtype in ['int64', 'float64']]
categorical_cols = [cname for cname in x.columns if x[cname].dtype == 'object']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer

# scaling 
numerical_transformer = Pipeline(steps = [
    ('scaler', RobustScaler())
])

# one-hot encoding
categorical_transformer = Pipeline(steps=[
    ('vectorizer', CountVectorizer(stop_words='english', ngram_range=(1,2), max_features=10)),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# bundle preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols),
      ])
```

### Q7] NaÃ¯ve Bayes Classification Algorithm on Sentiment category

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
pipe = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', model)])
```

```python
pipe.fit(x_train, y_train)
```

```python
pipe.score(x_test, y_test)
```

```python
y_pred = pipe.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

### Q8] Tabulate Accuracy

```python
tn = cm[0,0]
fp = cm[0,1]
fn = cm[1,0]
tp = cm[1,1]

precision = tp/(tp+fp)
recall = tp/(tp+fn)
f1  = 2*(precision*recall/(precision+recall))

metric = ['Precision', 'Recall', 'F1 score']
value = [precision, recall, f1]
d = {'Accuracy Metrics': metric, 'Values': value}
df3 = pd.DataFrame(data=d)
```
