```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import matplotlib.dates as mdates
import seaborn as sns

%matplotlib inline
```

```python
#Q1 Converted datestamp column to a date-time object and Set the date stamp columns as the index 
employment_df=pd.read_csv("employment.csv", parse_dates=True, index_col="datestamp")
employment_df.head()
```

```python
employment_df.info()
```

```python
#Q1 check if there are any missing values    
employment_df.isna().sum()
```

```python
print("2]Boxplot to find the distribution of unemployment rate for every industry")
fig, axes = pt.subplots(15, 1, figsize=(4, 50))
#fig, axes = pt.subplots(3, 5)
for name, ax in zip(['Agriculture', 'Business services', 'Construction', 'Durable goods manufacturing', 'Education and Health', 'Finance', 'Information', 'Leisure and hospitality', 'Manufacturing', 'Mining and Extraction', 'Nondurable goods manufacturing', 'Other', 'Self-employed', 'Transportation and Utilities', 'Wholesale and Retail Trade'], axes):
    sns.boxplot(data = employment_df, y=name, ax=ax)
    ax.set_ylabel("")
    ax.set_title(name)
    if ax != axes[-1]:
        ax.set_xlabel('')
```

```python
print("3]Unemployment rate of workers by industry using Line Chart")
employment_df.plot(subplots=True, figsize=(10, 20))
```

```python
df_year = employment_df.resample("Y").mean()
df_month = employment_df.resample("M").mean()
```

```python
print("4]Monthly and Yearly trends (Industry wise)")
fig, axes = pt.subplots(15, 1, figsize = (10,33), sharex = True)
for name, ax in zip(['Agriculture', 'Business services', 'Construction', 'Durable goods manufacturing', 'Education and Health', 'Finance', 'Information', 'Leisure and hospitality', 'Manufacturing', 'Mining and Extraction', 'Nondurable goods manufacturing', 'Other', 'Self-employed', 'Transportation and Utilities', 'Wholesale and Retail Trade'], axes):
    ax.plot(employment_df[name])
    ax.plot(df_year[name])
    ax.set_title(name)
```

```python
print("4]Monthly and Yearly trends")
df_year.plot(figsize=(17,10), title = 'Yearly')
df_month.plot(figsize=(17,10), title = 'Monthly')
```

```python
from pylab import rcParams
import statsmodels.api as sm
col_array = ['Agriculture', 'Business services', 'Construction', 'Durable goods manufacturing', 'Education and Health', 'Finance', 'Information', 'Leisure and hospitality', 'Manufacturing', 'Mining and Extraction', 'Nondurable goods manufacturing', 'Other', 'Self-employed', 'Transportation and Utilities', 'Wholesale and Retail Trade']
for col in col_array:
    rcParams['figure.figsize'] = 11, 9
    decomposition = sm.tsa.seasonal_decompose(employment_df[col], model='Additive')
    #The seasonal_decompose() method can take up to six parameters. 
    fig = decomposition.plot()
pt.show()
```

```python
print("5]Trend & Seasonality of Dataset:")
for col in col_array:
    decomposition = sm.tsa.seasonal_decompose(employment_df[col], model='Additive')
    fig = decomposition.trend.plot(title = 'Trend',figsize=(15,7), label = col, legend = True)
pt.show()
for col in col_array:
    decomposition = sm.tsa.seasonal_decompose(employment_df[col], model='Additive')
    fig =  decomposition.seasonal.plot(title = 'Seasonality',figsize=(15,7),  label = col , legend = True)
pt.show()
```

```python
print("6]Seasonality of Agriculture, Health and Finance sector.")
array = ['Agriculture','Education and Health', 'Finance']
for col in array:
    decomposition = sm.tsa.seasonal_decompose(employment_df[col], model='Additive')
    fig =  decomposition.seasonal.plot(title = 'Seasonality',figsize=(15,7),  label = col, legend = True)
    pt.show()
```

```python
print("7]Correlation between each time series in the dataset.")
sns.heatmap(employment_df.corr(), annot=True)
```
